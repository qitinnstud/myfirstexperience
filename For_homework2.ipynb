Импорт библиотеки pandas

```python
import pandas as pd
```

Читаю DataFrame c GitHub в переменную df и вывожу несколько строк

```python
#df.to_csv("drive/MyDrive/Weather_Data1.csv",sep='|',index=False)

url='https://raw.githubusercontent.com/qitinnstud/myfirstexperience/For_homework/Weather_Data.raw'
df = pd.read_csv(url, sep=',')
df.head()
```

Замена типа 'object' на 'datestamp'

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

Подсчет пустых значений

```python
df.isna().sum()
```

Всего пропусков:

```python
df.isna().sum().sum()
```

Сохраняем 'маску-пропуски' в DF x :

```python
x=df.isnull()
x.head()
```

Выводим строки в котором есть хоть один NaN :

```python
#df.loc[df.isnull()[:] == True] - не получилось...((
x['or']=x.astype('Int64').sum(axis=1).astype('bool') #Ну да, вопрос: как 'по-нормальному' векторизовать функцию OR в python?
df[x['or']!=False]
```

То же:

```python
# Нашел фукцию any ("Oh, how this all is not easy!"):
x['or']=x.any(axis=1)
df[x['or']==True]
```

Выводим корреляционную "матрицу":

```python
cr=df.corr()
cr.style.background_gradient(cmap='coolwarm')
```

Создаем "Топ 5" значений корреляции:

```python
y=[]
s=cr.shape[0]
for j in range(0,s-1):
    for i in range(j+1,s):
        yy=cr.loc[cr.index[i]][cr.columns[j]]
        y+=[[cr.index[i], cr.columns[j], yy, abs(yy)]]
y=pd.DataFrame(y)
y=y.sort_values(3, ascending=False)
y.head()
```
