Выводим корреляционную "матрицу":

```python
cr=df.corr()
cr.style.background_gradient(cmap='coolwarm')
```

Много хорошо скоррелированных величин.Создаем "Топ 5" значений корреляции:

```python
y=[]
s=cr.shape[0]
for j in range(0,s-1):
    for i in range(j+1,s):
        yy=cr.loc[cr.index[i]][cr.columns[j]]
        y+=[[cr.index[i], cr.columns[j], yy, abs(yy)]]
y=pd.DataFrame(y)
y=y.sort_values(3, ascending=False)
y.head()
```

Наблюдаем корреляции: прямые и обратные, от заметных и значительных до больших.

Случайная выборка из 20 значений:

```python
dfs=df.sample(20); dfs
```

```python
def frame(x, y):
    x=x[y]
    x=[x.min(), x.median(), round(x.mean(),3), x.max(), round(x.std(),3)]
    y=['min:', 'median:','mean:', 'max:', 'std:','value']
    dfx=pd.DataFrame(x, index=y[:5], columns=[y[5]])
    return dfx
```

```python
dfxt1=frame(dfs,'Temp_C')
dfxt1
```

```python
dfxt2=frame(dfs,'Dew Point Temp_C')
dfxt2
```

Операторы сравненеия:

```python
print(dfxt1 <= dfxt2)
```
