Подсчет всех строк с пустыми значениями и определение их процентного отношения к общему числу строк:

```python
r_na_count=df[df.isna().any(axis=1)== True].shape[0]
pr_na=round(r_na_count/df.shape[0] * 100, 1)
print(f'Число строк с пустыми значениями = {r_na_count} и процент пропусков = {pr_na}% ')
```

Обоснование удаления пропусков:

```python
if pr_na<=5:
    print(f'Процент пропусков {pr_na}% не более 5% => Можно удалить')
else:
    print(f'Процент пропусков {pr_na}% более 5% => Заполнить')
```
