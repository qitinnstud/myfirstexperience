Импорт библиотеки pandas

```python
import pandas as pd
```

Читаю DataFrame c GitHub в переменную df

```python
#df.to_csv("drive/MyDrive/Weather_Data1.csv",sep='|',index=False)

url='https://raw.githubusercontent.com/qitinnstud/myfirstexperience/For_homework/Weather_Data.raw'
df = pd.read_csv(url, sep=',')
df
```

Замена типа 'object' на 'datestamp'

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df['Date/Time'].sample().dtype=='datetime64[ns]'
```

Поиск и вывод дублирующихся строк:

```python
x=df.duplicated()
df[x==True]
```

Дубликаты строк отсутствуют:

```python
print(f'Число дублирующихся строк = {x[x==True].count()}')
```

Разделение датафрейма: сделаем выборки из исходного датафрейма:

- первый датафрейм

```python
s=df.shape[1]
df_l=df.iloc[:16,:s-3]; df_l
```

- второй датафрейм

```python
df_r=df.iloc[11:21,[0] + list(range(s-3,s))]; df_r
```

Объединение новых таблиц разными способами:

```python
# - способ 1
res = pd.merge(df_l, df_r, how='left', on='Date/Time');res
```

```python
# - способ 2
res = pd.merge(df_l, df_r, how='right', on='Date/Time');res
```

```python
# - способ 3
res = pd.merge(df_l, df_r, how='inner');res
```

```python
# - способ 4
res = pd.merge(df_l, df_r, how='outer');res
```

Подсчет всех строк с пустыми значениями и определение их процентного отношения к общему числу строк:

```python
r_na_count=df[df.isna().any(axis=1)== True].shape[0]
pr_na=round(r_na_count/df.shape[0] * 100, 1)
print(f'Число строк с пустыми значениями = {r_na_count} и процент пропусков = {pr_na}% ')
```

Обоснование удаления пропусков:

```python
if pr_na<=5:
    print(f'Процент пропусков {pr_na}% не более 5% => Можно удалить')
else:
    print(f'Процент пропусков {pr_na}% более 5% => Заполнить')
```
