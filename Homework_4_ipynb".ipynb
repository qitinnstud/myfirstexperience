Вычислить минимальную и масимальную площадь новой и вторичной недвижимости

```python
df.groupby('Transaction')['Area'].min()
```

```python
df.groupby('Transaction')['Area'].max()
```

```python
df.groupby('Transaction')['Area'].agg(['min', 'max'])
```

Найти среднюю площадь у недвижисти в зависимоти от кол-ва комнат

```python
df.groupby('BHK')['Area'].mean().apply(lambda x: round(x, 0))
```

Сколько в среденем стоят квартиры в зависимости от степени готовности и типа жилья среди новой недвижимости

```python
df[df['Transaction'] == 'New_Property'].groupby(['Status', 'Type'])['Price'].mean().reset_index()
```
